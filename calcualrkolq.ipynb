```python
import ipywidgets as widgets
from IPython.display import display

# Função para calcular a expressão
def calculate_expression(expression):
    try:
        return str(eval(expression))
    except:
        return "Erro"

# Função que atualiza o visor
def on_button_click(b):
    global expression
    if b.description == "=":
        result = calculate_expression(expression)
        display_text.value = result
        expression = result
    elif b.description == "C":
        expression = ""
        display_text.value = expression
    else:
        expression += b.description
        display_text.value = expression

# Variável para armazenar a expressão
expression = ""

# Criando o visor da calculadora
display_text = widgets.Text(value="", disabled=True, layout=widgets.Layout(width="400px", height="50px"), 
                            style={'description_width': 'initial'})

# Botões da calculadora
buttons = [
    '7', '8', '9', '/', 
    '4', '5', '6', '*', 
    '1', '2', '3', '-', 
    '0', '.', 'C', '+', 
    '='
]

# Estilizando os botões com rosa choque
button_style = {
    'button_color': '#FF69B4',  # Cor rosa choque
    'font_weight': 'bold',      # Negrito
    'width': '90px',
    'height': '50px',
    'border': '2px solid #FF1493'  # Borda rosa mais escura
}

# Criando os botões e associando a função a eles
button_widgets = []
for button in buttons:
    b = widgets.Button(description=button, layout=widgets.Layout(width="90px", height="50px"))
    b.style.button_color = button_style['button_color']
    b.style.font_weight = button_style['font_weight']
    b.layout.border = button_style['border']
    button_widgets.append(b)
    b.on_click(on_button_click)

# Layout dos botões
grid = widgets.GridBox(
    button_widgets,
    layout=widgets.Layout(
        grid_template_columns="repeat(4, 100px)",
        grid_template_rows="repeat(5, 50px)",
        grid_gap="10px"
    )
)

# Exibindo a calculadora
display(display_text, grid)
```
